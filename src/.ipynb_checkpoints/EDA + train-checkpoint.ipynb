# EDA + Train (Safety Recall Focus)

Цель проекта: **safety screener** для класса `p` (ядовитые грибы).
Ключевая задача: минимизировать FN по классу `p` через высокий `recall_p`.


## 1. EDA часть

## 2. Обучение логистической регрессии (baseline)


```python
DROP_COLUMNS = [
    "cap-surface", "gill-attachment", "gill-spacing", "stem-root",
    "stem-surface", "veil-type", "veil-color", "spore-print-color",
]

NUM_FEATURES = ["cap-diameter", "stem-height", "stem-width"]
CAT_FEATURES = [
    "cap-shape", "cap-color", "does-bruise-or-bleed", "gill-color",
    "stem-color", "ring-type", "habitat", "season", "has-ring",
]

OUTLIER_MAX = {"cap-diameter": 57, "stem-height": 30, "stem-width": 60}
required_features = NUM_FEATURES + CAT_FEATURES
```

```python
def apply_cleaning(df_in: pd.DataFrame, *, is_train: bool) -> pd.DataFrame:
    out = df_in.copy()
    out = out.drop(columns=[c for c in DROP_COLUMNS if c in out.columns])

    if is_train:
        for col, max_v in OUTLIER_MAX.items():
            if col in out.columns:
                out = out[out[col] <= max_v]

    return out


def ensure_contract(df_in: pd.DataFrame, features: list[str]) -> pd.DataFrame:
    out = df_in.copy()
    for c in features:
        if c not in out.columns:
            out[c] = np.nan
    return out[features]


def prepare_train_test(train_df: pd.DataFrame, test_df: pd.DataFrame):
    train_clean = apply_cleaning(train_df, is_train=True)
    test_clean = apply_cleaning(test_df, is_train=False)

    test_ids = test_clean["id"].copy() if "id" in test_clean.columns else None

    train_clean = train_clean.drop(columns=["id"], errors="ignore")
    test_clean = test_clean.drop(columns=["id"], errors="ignore")

    train_clean = train_clean[["class"] + [c for c in required_features if c in train_clean.columns]]
    test_clean = ensure_contract(test_clean, required_features)

    return train_clean, test_clean, test_ids
```

```python
test_raw = pd.read_csv(TEST_PATH)
train_clean, test_clean, test_ids = prepare_train_test(df.reset_index(), test_raw)

print("train_clean:", train_clean.shape)
print("test_clean:", test_clean.shape)
```

```python
# Save processed data once (parquet with CSV fallback)
try:
    train_clean.to_parquet(PROCESSED_DIR / "train_clean.parquet", index=False)
    test_clean.to_parquet(PROCESSED_DIR / "test_clean.parquet", index=False)
    print("Saved parquet to", PROCESSED_DIR)
except Exception as exc:
    print("Parquet unavailable, fallback to CSV:", type(exc).__name__)
    train_clean.to_csv(PROCESSED_DIR / "train_clean.csv", index=False)
    test_clean.to_csv(PROCESSED_DIR / "test_clean.csv", index=False)
```

```python
X = train_clean.drop(columns=["class"])
y = train_clean["class"]

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

num_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("log", FunctionTransformer(np.log1p, validate=False)),
    ("scaler", StandardScaler()),
])

# Important: constant imputation avoids fold failures on sparse/empty categories
cat_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="constant", fill_value="missing")),
    ("ohe", OneHotEncoder(handle_unknown="ignore")),
])

transformer = ColumnTransformer([
    ("num", num_pipe, NUM_FEATURES),
    ("cat", cat_pipe, CAT_FEATURES),
], remainder="drop")

logreg = LogisticRegression(
    solver="saga",
    C=0.5,
    class_weight={"e": 1, "p": 2.5},
    max_iter=2000,
    n_jobs=-1,
    random_state=42,
)

baseline_model = Pipeline([
    ("transformer", transformer),
    ("model", logreg),
])

baseline_model.fit(X_train, y_train)
baseline_proba = baseline_model.predict_proba(X_val)[:, list(baseline_model.classes_).index("p")]
```

```python
def find_threshold_for_recall(y_true_bin: np.ndarray, proba: np.ndarray, target: float = 0.99):
    rows = []
    for t in np.linspace(0.0, 1.0, 1001):
        pred = (proba >= t).astype(int)
        rows.append({
            "threshold": float(t),
            "recall_p": recall_score(y_true_bin, pred, zero_division=0),
            "precision_p": precision_score(y_true_bin, pred, zero_division=0),
            "accuracy": accuracy_score(y_true_bin, pred),
            "fn": int(((y_true_bin == 1) & (pred == 0)).sum()),
        })

    thr_df = pd.DataFrame(rows)
    feasible = thr_df[thr_df["recall_p"] >= target]

    if feasible.empty:
        best_t = 0.5
    else:
        best_t = float(feasible["threshold"].min())

    return best_t, thr_df
```

```python
y_val_bin = (y_val.astype(str) == "p").astype(int)
baseline_threshold_p, baseline_thr_df = find_threshold_for_recall(y_val_bin, baseline_proba, target=0.99)
baseline_pred = (baseline_proba >= baseline_threshold_p).astype(int)

baseline_auc = roc_auc_score(y_val_bin, baseline_proba)
baseline_recall = recall_score(y_val_bin, baseline_pred, zero_division=0)
baseline_precision = precision_score(y_val_bin, baseline_pred, zero_division=0)
baseline_acc = accuracy_score(y_val_bin, baseline_pred)

print("Baseline AUC:", round(baseline_auc, 5))
print("Baseline threshold_p:", round(baseline_threshold_p, 4))
print("Baseline recall_p:", round(baseline_recall, 5))
print("Baseline precision_p:", round(baseline_precision, 5))
```

## 3. CatBoost + safety threshold + Optuna


```python
X_train_cb, X_val_cb, y_train_cb, y_val_cb = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

y_tr = (y_train_cb.astype(str) == "p").astype(int)
y_va = (y_val_cb.astype(str) == "p").astype(int)

cat_cols = [c for c in CAT_FEATURES if c in X_train_cb.columns]
cat_idx = [X_train_cb.columns.get_loc(c) for c in cat_cols]
nan_token = "__nan__"

X_tr = X_train_cb.copy()
X_va = X_val_cb.copy()
X_tr[NUM_FEATURES] = np.log1p(X_tr[NUM_FEATURES])
X_va[NUM_FEATURES] = np.log1p(X_va[NUM_FEATURES])
X_tr[cat_cols] = X_tr[cat_cols].fillna(nan_token).astype(str)
X_va[cat_cols] = X_va[cat_cols].fillna(nan_token).astype(str)
```

```python
def build_catboost(params: dict):
    return CatBoostClassifier(
        loss_function="Logloss",
        eval_metric="AUC",
        iterations=int(params.get("iterations", 600)),
        learning_rate=float(params.get("learning_rate", 0.06)),
        depth=int(params.get("depth", 8)),
        l2_leaf_reg=float(params.get("l2_leaf_reg", 8.0)),
        random_strength=float(params.get("random_strength", 1.0)),
        bootstrap_type="Bernoulli",
        subsample=float(params.get("subsample", 0.8)),
        od_type="Iter",
        od_wait=100,
        verbose=False,
        allow_writing_files=False,
    )
```

```python
def evaluate_catboost_with_threshold(params: dict, target_recall: float = 0.99):
    model = build_catboost(params)
    train_pool = Pool(X_tr, y_tr, cat_features=cat_idx)
    val_pool = Pool(X_va, y_va, cat_features=cat_idx)

    model.fit(train_pool, eval_set=val_pool, use_best_model=True)
    proba = model.predict_proba(val_pool)[:, 1]

    threshold_p, _ = find_threshold_for_recall(y_va, proba, target=target_recall)
    pred = (proba >= threshold_p).astype(int)

    metrics = {
        "auc": roc_auc_score(y_va, proba),
        "recall_p": recall_score(y_va, pred, zero_division=0),
        "precision_p": precision_score(y_va, pred, zero_division=0),
        "accuracy": accuracy_score(y_va, pred),
        "fn": int(((y_va == 1) & (pred == 0)).sum()),
        "threshold_p": threshold_p,
    }
    return model, proba, metrics
```

```python
base_params = {
    "iterations": 600,
    "learning_rate": 0.06,
    "depth": 8,
    "l2_leaf_reg": 8.0,
    "random_strength": 1.0,
    "subsample": 0.8,
}

cat_model, cat_val_proba, cat_metrics = evaluate_catboost_with_threshold(base_params, target_recall=0.99)
cat_metrics
```

```python
# Safety evaluation report
val_pred_selected = (cat_val_proba >= cat_metrics["threshold_p"]).astype(int)
cm = confusion_matrix(y_va, val_pred_selected)

report = pd.DataFrame([
    {
        "model": "LogReg baseline",
        "auc": baseline_auc,
        "recall_p": baseline_recall,
        "precision_p": baseline_precision,
        "accuracy": baseline_acc,
        "threshold_p": baseline_threshold_p,
    },
    {
        "model": "CatBoost",
        "auc": cat_metrics["auc"],
        "recall_p": cat_metrics["recall_p"],
        "precision_p": cat_metrics["precision_p"],
        "accuracy": cat_metrics["accuracy"],
        "threshold_p": cat_metrics["threshold_p"],
    },
])

report
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["e(0)", "p(1)"])
disp.plot(cmap="Blues")
plt.title(f"CatBoost @ threshold={cat_metrics['threshold_p']:.3f}")
plt.show()
```

```python
# Artifacts to keep as project contract
threshold_p = float(cat_metrics["threshold_p"])
review_band = (max(0.0, threshold_p - 0.05), threshold_p)

artifacts_contract = {
    "required_features": required_features,
    "threshold_p": threshold_p,
    "review_band": review_band,
}
artifacts_contract
```

```python
# Optional Optuna tuning (CV objective focused on recall constraint)
try:
    import optuna
    OPTUNA_AVAILABLE = True
except Exception as exc:
    OPTUNA_AVAILABLE = False
    print("Optuna is not installed:", type(exc).__name__)

OPTUNA_AVAILABLE
```

```python
if OPTUNA_AVAILABLE:
    import optuna
    from optuna.samplers import TPESampler
    from optuna.pruners import MedianPruner

    X_cv = X.copy()
    y_cv = (y.astype(str) == "p").astype(int)

    def objective(trial):
        params = {
            "depth": trial.suggest_int("depth", 6, 12),
            "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.15, log=True),
            "l2_leaf_reg": trial.suggest_float("l2_leaf_reg", 1.0, 20.0),
            "random_strength": trial.suggest_float("random_strength", 0.0, 3.0),
            "iterations": trial.suggest_int("iterations", 300, 1500),
            "subsample": trial.suggest_float("subsample", 0.6, 1.0),
        }

        skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
        fold_scores = []
        alpha = 3.0  # stronger FN penalty

        for fold, (tr_idx, va_idx) in enumerate(skf.split(X_cv, y_cv), start=1):
            X_fold_tr = X_cv.iloc[tr_idx].copy()
            X_fold_va = X_cv.iloc[va_idx].copy()
            y_fold_tr = y_cv.iloc[tr_idx].copy()
            y_fold_va = y_cv.iloc[va_idx].copy()

            fold_cat_cols = [c for c in CAT_FEATURES if c in X_fold_tr.columns]
            fold_cat_idx = [X_fold_tr.columns.get_loc(c) for c in fold_cat_cols]

            X_fold_tr[NUM_FEATURES] = np.log1p(X_fold_tr[NUM_FEATURES])
            X_fold_va[NUM_FEATURES] = np.log1p(X_fold_va[NUM_FEATURES])
            X_fold_tr[fold_cat_cols] = X_fold_tr[fold_cat_cols].fillna(nan_token).astype(str)
            X_fold_va[fold_cat_cols] = X_fold_va[fold_cat_cols].fillna(nan_token).astype(str)

            model = build_catboost(params)
            tr_pool = Pool(X_fold_tr, y_fold_tr, cat_features=fold_cat_idx)
            va_pool = Pool(X_fold_va, y_fold_va, cat_features=fold_cat_idx)
            model.fit(tr_pool, eval_set=va_pool, use_best_model=True)

            proba = model.predict_proba(va_pool)[:, 1]
            t_best, _ = find_threshold_for_recall(y_fold_va.values, proba, target=0.99)
            pred = (proba >= t_best).astype(int)

            recall_p = recall_score(y_fold_va, pred, zero_division=0)
            precision_p = precision_score(y_fold_va, pred, zero_division=0)
            fn_rate = (((y_fold_va == 1) & (pred == 0)).sum()) / max(1, (y_fold_va == 1).sum())

            if recall_p < 0.99:
                fold_score = -10.0 - (0.99 - recall_p) * 100.0
            else:
                fold_score = precision_p - alpha * fn_rate

            fold_scores.append(fold_score)
            trial.report(float(np.mean(fold_scores)), step=fold)
            if trial.should_prune():
                raise optuna.TrialPruned()

        return float(np.mean(fold_scores))

    study = optuna.create_study(
        direction="maximize",
        sampler=TPESampler(seed=42),
        pruner=MedianPruner(n_startup_trials=10),
        study_name="catboost_safety_recall",
    )

    # Budget target from plan: ~45-60 min
    study.optimize(objective, timeout=60*60, show_progress_bar=True)

    print("Best score:", study.best_value)
    print("Best params:", study.best_params)
else:
    print("Skip Optuna block. Install with: UV_CACHE_DIR=.uv-cache uv pip install optuna")
```

```python
if OPTUNA_AVAILABLE and 'study' in globals() and len(study.trials) > 0:
    best_params = study.best_params
else:
    best_params = base_params

final_model, final_val_proba, final_metrics = evaluate_catboost_with_threshold(best_params, target_recall=0.99)
threshold_p = float(final_metrics["threshold_p"])
review_band = (max(0.0, threshold_p - 0.05), threshold_p)

final_metrics
```

## 4. Submission на тестовых данных для Kaggle


```python
X_t = ensure_contract(test_clean.copy(), required_features)
X_t[NUM_FEATURES] = np.log1p(X_t[NUM_FEATURES])
X_t[CAT_FEATURES] = X_t[CAT_FEATURES].fillna(nan_token).astype(str)

cat_idx_test = [X_t.columns.get_loc(c) for c in CAT_FEATURES if c in X_t.columns]
test_pool = Pool(X_t, cat_features=cat_idx_test)
test_proba = final_model.predict_proba(test_pool)[:, 1]
pred = np.where(test_proba >= threshold_p, "p", "e")

answers = pd.read_csv(SAMPLE_SUB_PATH, index_col="id")
answers["class"] = pred
answers.head()
```

```python
SUB_PATH = DATA_DIR / "submission.csv"
answers.to_csv(SUB_PATH)
print("Saved:", SUB_PATH)
```

## 5. Результат

- Primary metric: `recall_p` for class `p`
- Threshold policy: minimal threshold with `recall_p >= 0.99`
- Submission uses `predict_proba` + `threshold_p` (not default `predict`)
- Project artifacts: `required_features`, `threshold_p`, `review_band`
