# Mushroom Classification (Kaggle S4E8)

Single notebook workflow: EDA -> cleaning -> baseline -> CatBoost -> submission.


## 1. Problem & Data

- Competition: Kaggle Playground Series S4E8
- Target: `class` (`p` / `e`)
- Local files: `data/train.csv`, `data/test.csv`, `data/sample_submission.csv`


```python
from pathlib import Path

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, FunctionTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, accuracy_score

from catboost import Pool, CatBoostClassifier
```

```python
DATA_DIR = Path("data")
PROCESSED_DIR = DATA_DIR / "processed"
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

train_raw = pd.read_csv(DATA_DIR / "train.csv")
test_raw = pd.read_csv(DATA_DIR / "test.csv")

print("train shape:", train_raw.shape)
print("test shape:", test_raw.shape)
train_raw.head(3)
```

## 2. EDA

Quick checks: class balance, missing values, numeric distributions.


```python
train_raw["class"].value_counts(normalize=True).rename("ratio")
```

```python
missing_pct = (train_raw.isna().mean() * 100).sort_values(ascending=False)
missing_pct.head(12)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, col in zip(axes, ["cap-diameter", "stem-height", "stem-width"]):
    sns.histplot(train_raw[col], ax=ax, bins=60)
    ax.set_title(col)
plt.tight_layout()
```

## 3. Cleaning Rules

Rules fixed from EDA:
- drop high-missing columns
- remove extreme numeric outliers on train only
- keep feature contract (`required_features`) for future inference


```python
DROP_COLUMNS = [
    "cap-surface",
    "gill-attachment",
    "gill-spacing",
    "stem-root",
    "stem-surface",
    "veil-type",
    "veil-color",
    "spore-print-color",
]

NUM_FEATURES = ["cap-diameter", "stem-height", "stem-width"]
CAT_FEATURES = [
    "cap-shape",
    "cap-color",
    "does-bruise-or-bleed",
    "gill-color",
    "stem-color",
    "ring-type",
    "habitat",
    "season",
]
BINARY_FEATURES = ["has-ring"]

OUTLIER_MAX = {
    "cap-diameter": 57,
    "stem-height": 30,
    "stem-width": 60,
}

required_features = NUM_FEATURES + CAT_FEATURES + BINARY_FEATURES
```

```python
def apply_cleaning(df: pd.DataFrame, *, is_train: bool) -> pd.DataFrame:
    out = df.copy()

    drop_cols = [c for c in DROP_COLUMNS if c in out.columns]
    out = out.drop(columns=drop_cols)

    if is_train:
        for col, max_val in OUTLIER_MAX.items():
            if col in out.columns:
                out = out[out[col] <= max_val]

    return out


def ensure_contract(df: pd.DataFrame, features: list[str]) -> pd.DataFrame:
    # Keep only required columns and add missing ones as NaN.
    out = df.copy()
    for col in features:
        if col not in out.columns:
            out[col] = np.nan
    return out[features]
```

## 4. Save Processed Data


```python
train_clean = apply_cleaning(train_raw, is_train=True)
test_clean = apply_cleaning(test_raw, is_train=False)

train_clean = train_clean.drop(columns=["id"], errors="ignore")
test_ids = test_clean["id"].copy()
test_clean = test_clean.drop(columns=["id"], errors="ignore")

train_clean = train_clean[["class"] + [c for c in required_features if c in train_clean.columns]]
test_clean = ensure_contract(test_clean, required_features)

print("train_clean shape:", train_clean.shape)
print("test_clean shape:", test_clean.shape)
```

```python
# Prefer parquet. If parquet engine is unavailable, fallback to csv.
train_parquet = PROCESSED_DIR / "train_clean.parquet"
test_parquet = PROCESSED_DIR / "test_clean.parquet"

try:
    train_clean.to_parquet(train_parquet, index=False)
    test_clean.to_parquet(test_parquet, index=False)
    print("saved:", train_parquet)
    print("saved:", test_parquet)
except Exception as exc:
    print("parquet save failed, fallback to csv. Install pyarrow to use parquet.")
    print(type(exc).__name__, str(exc))
    train_clean.to_csv(PROCESSED_DIR / "train_clean.csv", index=False)
    test_clean.to_csv(PROCESSED_DIR / "test_clean.csv", index=False)
```

## 5. Baseline (LogReg, short)


```python
df_model = train_clean.copy()
X = df_model.drop(columns=["class"])
y = df_model["class"]

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

num_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("log", FunctionTransformer(np.log1p, validate=False)),
    ("scaler", StandardScaler()),
])

cat_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore")),
])

bin_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore")),
])

preproc = ColumnTransformer([
    ("num", num_pipe, NUM_FEATURES),
    ("cat", cat_pipe, CAT_FEATURES),
    ("bin", bin_pipe, BINARY_FEATURES),
], remainder="drop")

logreg = Pipeline([
    ("preproc", preproc),
    ("model", LogisticRegression(max_iter=300)),
])

logreg.fit(X_train, y_train)
logreg_proba = logreg.predict_proba(X_val)[:, list(logreg.classes_).index("p")]
logreg_pred = logreg.predict(X_val)

logreg_auc = roc_auc_score((y_val == "p").astype(int), logreg_proba)
logreg_acc = accuracy_score(y_val, logreg_pred)

print("LogReg AUC:", round(logreg_auc, 5))
print("LogReg ACC:", round(logreg_acc, 5))
```

## 6. CatBoost (main model)


```python
X_train_cb, X_val_cb, y_train_cb, y_val_cb = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

y_tr = (y_train_cb.astype(str) == "p").astype(int)
y_va = (y_val_cb.astype(str) == "p").astype(int)

cat_cols = CAT_FEATURES + BINARY_FEATURES
cat_idx = [X_train_cb.columns.get_loc(c) for c in cat_cols]
nan_token = "__nan__"

X_tr = X_train_cb.copy()
X_va = X_val_cb.copy()

X_tr[NUM_FEATURES] = np.log1p(X_tr[NUM_FEATURES])
X_va[NUM_FEATURES] = np.log1p(X_va[NUM_FEATURES])
X_tr[cat_cols] = X_tr[cat_cols].fillna(nan_token).astype(str)
X_va[cat_cols] = X_va[cat_cols].fillna(nan_token).astype(str)

train_pool = Pool(X_tr, y_tr, cat_features=cat_idx)
val_pool = Pool(X_va, y_va, cat_features=cat_idx)

cat_model = CatBoostClassifier(
    loss_function="Logloss",
    eval_metric="AUC",
    iterations=200,
    learning_rate=0.08,
    depth=10,
    l2_leaf_reg=8,
    random_strength=1.5,
    od_type="Iter",
    od_wait=50,
    verbose=20,
    allow_writing_files=False,
    bootstrap_type="Bernoulli",
    subsample=0.8,
)

cat_model.fit(train_pool, eval_set=val_pool, use_best_model=True)

cat_proba = cat_model.predict_proba(val_pool)[:, 1]
cat_pred = (cat_proba >= 0.5).astype(int)

cat_auc = roc_auc_score(y_va, cat_proba)
cat_acc = accuracy_score(y_va, cat_pred)

print("CatBoost AUC:", round(cat_auc, 5))
print("CatBoost ACC:", round(cat_acc, 5))
```

```python
pd.DataFrame(
    [
        {"model": "LogReg", "val_auc": logreg_auc, "val_acc": logreg_acc},
        {"model": "CatBoost", "val_auc": cat_auc, "val_acc": cat_acc},
    ]
).sort_values("val_auc", ascending=False)
```

## 7. Test Prediction & Submission


```python
X_t = test_clean.copy()
X_t = ensure_contract(X_t, required_features)

X_t[NUM_FEATURES] = np.log1p(X_t[NUM_FEATURES])
X_t[cat_cols] = X_t[cat_cols].fillna(nan_token).astype(str)

test_pool = Pool(X_t, cat_features=cat_idx)

test_proba = cat_model.predict_proba(test_pool)[:, 1]
pred_labels = np.where(test_proba >= 0.5, "p", "e")

submission = pd.DataFrame({"id": test_ids, "class": pred_labels})
submission.to_csv(DATA_DIR / "submission.csv", index=False)

submission.head()
```
